```python
# Matplotlib
# Matplotlib is a comprehensive library for creating static, animated,
# and interactive visualizations in Python.
# Matplotlib makes easy things easy and hard things possible.
 
# You can generate plots, histograms, power spectra, bar charts, errorcharts,
# scatterplots, etc., with just a few lines of code.
 
# The matplotlib library contains several modules. The matplotlib.pyplot module is
# designed to create colorful plots.
#has context menu
import matplotlib.pyplot as plt
```

```python
#Key Features:
#Versatile plotting functions (line, bar, scatter, histogram, pie, 3D).
#Highly customizable plots, supporting detailed styling and annotations.
#Works seamlessly with NumPy and Pandas data.
#Cross-platform compatibility for desktop and web environments.

#Use Cases:
#Exploratory and scientific data analysis.
#Statistical visualization and reporting.
#Visualizing machine learning/model results.
#Publication-quality charts for presentations and research.
#Time series and financial data visualization
```

```python
#Matplotlib Graph Types
#1. Line Plot
#Use Case: Time series or trend visualization

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]

plt.plot(x, y, marker='o')
plt.title("Line Plot Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()

#2. Bar Chart
#Use Case: Comparing categorical data

categories = ['A', 'B', 'C', 'D']
values = [23, 45, 56, 78]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart Example")
plt.show()

#3. Scatter Plot
#Use Case: Analyzing correlation between variables

import numpy as np

x = np.random.rand(50)
y = x + np.random.normal(0, 0.1, 50)

plt.scatter(x, y, color='green')
plt.title("Scatter Plot Example")
plt.show()

 #4. Histogram
#Use Case: Distribution of numerical data

data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple')
plt.title("Histogram Example")
plt.show()

#5. Pie Chart
#Use Case: Show proportions of a whole

labels = ['A', 'B', 'C']
sizes = [40, 35, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Pie Chart Example")
plt.show()
```
