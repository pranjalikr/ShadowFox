```python
#Seaborn
#Seaborn is Matplotlib’s stylish cousin, made to create beautiful charts with less code. It automatically calculates and shows helpful statistics—like color-coded heatmaps and distribution plots—which makes it awesome for revealing patterns quickly.

#Key strengths:
#Gorgeous default styles and smart color options
#Plots statistical relationships and comparisons easily

#How it’s used:
#Exploring relationships between variables
#Visualizing how data clusters or spreads
#Quickly spotting correlations and patterns
```

```python
#Seaborn Graph Types
#1. Line Plot
#Use Case: Similar to Matplotlib, but works well with Pandas

import seaborn as sns
import pandas as pd

df = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [5, 10, 8, 12, 15]
})

sns.lineplot(data=df, x="x", y="y")
```

```python
#2. Bar Plot
#Use Case: Automatically aggregates and visualizes means

df = pd.DataFrame({
    "Category": ["A", "B", "C", "A", "B", "C"],
    "Value": [4, 7, 1, 5, 6, 2]
})

sns.barplot(x="Category", y="Value", data=df)
```

```python
#3. Scatter Plot (relplot)
#Use Case: Relational plots across multiple categories

tips = sns.load_dataset("tips")

sns.relplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
#4. Histogram (distplot / displot)
#Use Case: View distribution of a single variable

sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
```

```python
#5. Heatmap
#Use Case: Correlation matrix or grid data
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

corr = tips.select_dtypes(include='number').corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```
